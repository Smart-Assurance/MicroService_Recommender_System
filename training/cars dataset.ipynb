```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## dataset new cars

## dataset cars global

```python
df = pd.concat([df1, df2], ignore_index=True)
```

```python
df.head()
```

```python
df.to_csv('global_cars.csv', index=False, encoding='utf-8')
```

### data preprocessing

```python
df.info()
```

```python
# missing values 
sns.heatmap(df.isnull(),cmap = 'magma',cbar = False) 
```

```python
# % of nan value of each columns  
for i in df.columns:
    null_rate = df[i].isnull().sum() / len(df)*100
    print("{}'s null rate: {}%".format(i, round(null_rate, 2)))
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df.info()
```

```python
# Remove rows with NaN values
df = df.dropna()
df.info()
```

```python
# missing values 
sns.heatmap(df.isnull(),cmap = 'magma',cbar = False) 
```

### Marques visualisation

```python
df['marque'].unique()
```

```python
# Count the occurrences of each marque
marque_counts = df['marque'].value_counts()

plt.figure(figsize=(15, 6))
# Plot the counts using a bar plot
plt.bar(marque_counts.index, marque_counts.values)
plt.xticks(rotation='vertical')
plt.xlabel('Marque')
plt.ylabel('Count')
plt.title('Count of each Marque')
plt.show()
```

```python
marque_counts = df['marque'].value_counts()

# Increase the size of the figure
plt.figure(figsize=(15,12))

# Plot the counts using a horizontal bar plot
bars = plt.barh(marque_counts.index, marque_counts.values)

# Annotate each bar with its count
for bar, count in zip(bars, marque_counts.values):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, str(count), ha='left', va='center')

plt.xlabel('Count')
plt.ylabel('Marque')
plt.title('Count of each Marque')
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
# Group by 'marque' and calculate the mean price for each group
mean_prices = df.groupby('marque')['price'].mean()
plt.figure(figsize=(15, 6))
# Plot the counts using a bar plot
plt.bar(mean_prices.index, mean_prices.values)
plt.xticks(rotation='vertical')# Plot the mean prices using a bar graph
plt.xlabel('Marque')
plt.ylabel('Mean Price')
plt.title('Mean Price for Each Marque')
plt.gca().get_yaxis().set_major_formatter(FuncFormatter(lambda x, _: int(x)))
plt.show()
```

```python
# Exclude 'ferrari' from the DataFrame
df_excluding_ferrari = df[df['marque'] != 'ferrari']

# Group by 'marque' and calculate the mean price for each group
mean_prices = df_excluding_ferrari.groupby('marque')['price'].mean()
plt.figure(figsize=(15, 6))
# Plot the counts using a bar plot
plt.bar(mean_prices.index, mean_prices.values)
plt.xticks(rotation='vertical')# Plot the mean prices using a bar graph
plt.xlabel('Marque')
plt.ylabel('Mean Price')
plt.title('Mean Price for Each Marque')
plt.gca().get_yaxis().set_major_formatter(FuncFormatter(lambda x, _: int(x)))
plt.show()
```

```python
import json
marque_price_dict = dict(zip(mean_prices.index, mean_prices.values))

json_file_path = 'marque_price_dict.json'

# Save the dictionary to a JSON file
with open(json_file_path, 'w') as json_file:
    json.dump(marque_price_dict, json_file)

print(f"The dictionary has been successfully saved to {json_file_path}.")
```
