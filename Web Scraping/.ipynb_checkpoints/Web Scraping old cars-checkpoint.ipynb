```python
import json
from urllib.request import urlopen,Request
from bs4 import BeautifulSoup
import ssl
import urllib
import re

def extract_url(url):
    
    # Ignore SSL certificate errors
    ctx = ssl.create_default_context()
    ctx.check_hostname = False
    ctx.verify_mode = ssl.CERT_NONE

    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})

    html = urlopen(req, context=ctx).read()
    soup = BeautifulSoup(html, "html.parser")

    links = soup.find_all('div', class_='content-inner')
    
    data_list=[]
    for link in links:
        if link.find('div', class_='price color_primary PriceListing'):
            a=link.find('a')
            data=a['href']
            data_list.append(data)
    return data_list
```

```python
url='https://www.moteur.ma/fr/voiture/achat-voiture-occasion/'
results=extract_url(url)
for i in range(30, 27841, 30):
    results.extend(extract_url(url+str(i)))
    print(i)
```

```python
json_file_path = "old_cars_list.json"
with open(json_file_path, "w", encoding='utf-8') as json_file:
    json.dump(results, json_file, ensure_ascii=False, indent=4)

    print(f"The data has been successfully exported to {json_file_path}.")   
```

```python
import json
from urllib.request import urlopen,Request
from bs4 import BeautifulSoup
import ssl
import urllib
import re

def extract_info(url):
    
    # Ignore SSL certificate errors
    ctx = ssl.create_default_context()
    ctx.check_hostname = False
    ctx.verify_mode = ssl.CERT_NONE

    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})

    html = urlopen(req, context=ctx).read()
    soup = BeautifulSoup(html, "html.parser")
    
    marque_span = soup.find('span', class_='text_bold')
    price_div = soup.find('div', class_='color_primary text_bold price-block')
    detail_lines = soup.find_all('div', class_='detail_line')

    marque_span=marque_span.text.strip()
    marque=marque_span.split(' ')[0]
    model=marque_span.split(' ')[1]
    price = int(re.sub(r'\D', '', price_div.text.strip()))
    # Creating the dictionary
    car_info_dict = {
        'marque': marque,
        'model': model,  # You can set the model to None or any default value if not available in the HTML
        'price': price
    }

    # Iterate through detail_line divs and extract information
    for detail_line in detail_lines:
        key = detail_line.find('span', class_='col-md-6 col-xs-6').text.strip()
        value = detail_line.find('span', class_='text_bold').text.strip()
        car_info_dict[key] = value

    return car_info_dict
```

```python
extract_info('https://www.moteur.ma/fr/voiture/achat-voiture-occasion/detail-annonce/434385/dacia-sandero-.html')
```

```python
file_path = 'old_cars_list.json'

# Read the JSON file
with open(file_path, 'r') as file:
    urls = json.load(file)
```

```python
json_file_path = "old_cars.json"

for i in range (18695,len(urls)):
    car_info = extract_info(urls[i])
        
    # Write the result to the JSON file
    with open(json_file_path, "a", encoding='utf-8') as json_file:
        json.dump(car_info, json_file, ensure_ascii=False, indent=4)
        json_file.write(',\n')  # Add a newline between entries
    
    if i % 100 == 0:
        print(f"Processed {i} URLs")

print(f"The data has been successfully exported to {json_file_path}.")
```

```python
import json
import csv

# Assuming the JSON file is named 'old_cars.json'
json_file_path = 'old_cars.json'

# Read the JSON file
with open(json_file_path, 'r', encoding='utf-8') as json_file:
    cars_list = json.load(json_file)

# Specify the attributes to include in the CSV file
selected_attributes = ["marque", "model", "Puissance fiscale", "Carburant", "Année", "Boite de vitesses","price"]

# Specify the CSV file path
csv_file_path = "old_cars.csv"

# Write the header to the CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=selected_attributes)
    csv_writer.writeheader()

    # Write each car's information to the CSV file
    for car_info in cars_list:
        selected_info = {attr: car_info.get(attr, "") for attr in selected_attributes}
        csv_writer.writerow(selected_info)

print(f"The data has been successfully exported to {csv_file_path}.")
```
